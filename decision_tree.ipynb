```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data =load_breast_cancer()
dataset =pd.DataFrame(data=data['data'],columns=data['feature_names'])
dataset
```

```python
from sklearn.model_selection import train_test_split
X=dataset.copy()
y=data['target']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.30,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(max_depth=4)
clf=clf.fit(X_train,y_train)
```

```python
clf.get_params()
```

```python
predication=clf.predict(X_test)
predication
```

```python
clf.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predication)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predication,labels=[0,1])
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,predication)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, predication)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predication,target_names=['malignant','benign'] ))
```
